```python
viewport = np.array([96,96])
h,w = viewport
col, row = np.meshgrid(np.linspace(-1,1,h),np.linspace(-1,1,w))
points = jnp.stack((col,row), axis=2).reshape(h,w,2) 

camera = Camera(get_points_from_angles(0,0,2.1),jnp.array([0.,1.,0.]),jnp.array([0.,0.,0]))

light = Light(get_points_from_angles(20,0,1.7),1.0)
light_ref = Light(get_points_from_angles(30,30,2.0),1.0)

sphere = load_obj('obj/sphere_642.obj',True)
mesh_sphere = sphere[0][sphere[1].astype(int)]
mesh_sphere = jnp.concatenate([mesh_sphere,jnp.ones([mesh_sphere.shape[0],3,1])],axis=-1)
normals_sphere = sphere[3][sphere[1]]
normals_sphere = normalize(normals_sphere,-1)

material_sphere = Material(jnp.array([[0.2,0.2,0.2],[0.7,0.5,0.5],[0.9,0.9,0.9]]),100.)

obj1 = Object(mesh_sphere,normals_sphere)
objects = [obj1,]
materials = [material_sphere,]
```

```python
blurriness = -6.0
shading_mode = 'phong'
kernel = 'gaussian'
img_init = forward(points,objects,materials,light,camera,shading_mode,kernel,blurriness)
img_ref = forward(points,objects,materials,light_ref,camera,shading_mode,kernel,blurriness)
```

# single learn

```python
def training(points,objects,materials,light,camera,img_ref,shading_mode,kernel,blurriness,iteration):
    history = []
    history_para = []
    history_img = []

    beta1 = 0.9
    beta2 = 0.999
    eps = 1e-8
    lr = 0.1

    m = jnp.zeros_like(light.position)
    v = jnp.zeros_like(light.position)
    for i in trange(iteration):
        (loss,img),g = value_and_grad(mse_loss,argnums=(3),has_aux=True)(points,objects,materials,light,camera,img_ref,shading_mode,kernel,blurriness)

        history.append(loss)
        history_img.append(img)

        m = beta1*m + (1-beta1)*g.position
        v = beta2*v + (1-beta2)*jnp.square(g.position)
        m_hat = m/(1-jnp.asarray(beta1,m.dtype)**(i+1))
        v_hat = v/(1-jnp.asarray(beta2,m.dtype)**(i+1))
        light.position -= lr*m_hat/(jnp.sqrt(v_hat)+eps)
        
    return history,history_para,history_img
```

```python
start = time.time()
history,history_para,history_img = training(points,objects,materials,light,camera,img_ref,'phong','gaussian',blurriness,30)
end = time.time()
time_cost = end - start
```

```python
plt.plot(history)
plt.title('Time cost: %.5fs Final loss: %.5f' % (time_cost,history[-1]))
plt.xlabel('Iteration')
plt.ylabel('Sum of the difference between two images')
```
